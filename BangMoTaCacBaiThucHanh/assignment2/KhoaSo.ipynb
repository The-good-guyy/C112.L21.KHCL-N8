# **1.ABSTRACTION**

**INPUT**: một xâu ký tự số có độ lớn từ 2 đến $10^5$

**OUTPUT**: một xâu ký tự có giá trị lớn nhất được sắp xếp từ xâu ký tự cho trước và phải chia hết cho 3(có thể xóa bất cứ ký tự nào nếu muốn)

# **2.PATTERN RECOGNITION**

Kĩ thuật áp dụng: Xử lí chuỗi

Đặc điểm nhận dạng: Thay đổi vị trí hoặc xoá các vị trí trong chuỗi để được chuỗi có giá lớn nhất và chia hết cho 3

# **3.ALGORITHM DESIGNED**

Một chuỗi chia hết cho 3 khi tổng các chữ số trong chúng chia hết.

Nếu chúng đã chia hết cho 3 thì tốt bỏ qua bước này, còn không thì sẽ có 2 trường hợp:

Trường hợp 1:


1.   Chuỗi chia 3 dư 1:

*   Xóa 1 chữ số trong chuỗi là 1 , 4 , 7.
*   Xóa 2 chữ số trong chuỗi là 2 , 5 , 8.


2.   Chuỗi chia 3 dư 2:

*   Xóa 1 chữ số trong chuỗi là 2 , 5 , 8.

*   Xóa 2 chữ số trong chuỗi là 1 , 4 , 7.





```python
from sys import stdin, stdout
from collections import Counter
```

```python
n = input().strip()
count = Counter(n)
thua = sum(map(int, n)) % 3

if thua!=0:
    odd=['1','4','7']
    even=['2','5','8']
    odd = [x for x in odd if x in count]
    even= [x for x in even if x in count]
    if thua==1:
        if len(odd)==0:
            c = 2
            for x in even:
                t = min(c, count[x])
                c -= t
                count[x] -= t
        else:
            c=1
            for x in odd:
                t = min(c, count[x])
                c -= t
                count[x] -= t
    else:
        if len(even)==0:
            c = 2
            for x in odd:
                t = min(c, count[str(x)])
                c -= t
                count[x] -= t
        else:
            c=1
            for x in even:
                t = min(c, count[x])
                c -= t
                count[x] -= t


for x in count:
    stdout.write(x * count[x])
print()
```

# **PROGRAMMING**

```python
from sys import stdin, stdout
from collections import Counter

n = input().strip()
count = Counter(n)
thua = sum(map(int, n)) % 3

if thua!=0:
    odd=['1','4','7']
    even=['2','5','8']
    odd = [x for x in odd if x in count]
    even= [x for x in even if x in count]
    if thua==1:
        if len(odd)==0:
            c = 2
            for x in even:
                t = min(c, count[x])
                c -= t
                count[x] -= t
        else:
            c=1
            for x in odd:
                t = min(c, count[x])
                c -= t
                count[x] -= t
    else:
        if len(even)==0:
            c = 2
            for x in odd:
                t = min(c, count[str(x)])
                c -= t
                count[x] -= t
        else:
            c=1
            for x in even:
                t = min(c, count[x])
                c -= t
                count[x] -= t


for x in sorted(count.keys(), reverse=True):
    stdout.write(x * count[x])
print()
```
