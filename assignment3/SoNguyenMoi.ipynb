# **1. ABSTRACTION**

Cho số nguyên dương n có không quá 100 chữ số. Tìm số nguyên lớn nhất m chia hết cho 3 và khác n ở đúng một chữ số

# **2. PATTERN RECOGNITION**

Đây là bài toán số học **Tìm số lớn nhất chia hết cho 3 áp dụng tính chất tổng các phần tử chia hết cho 3.** Áp dụng kỹ thuật xử lý chuỗi

# **3. ALGORITHM DESIGNED**


*   **Bước 1**: Đưa từng chữ số của $n$ vào mảng $Num$ sao cho giữ nguyên thứ tự của các chữ số $n$


*   **Bước 2**: Tính $add$ và $sub$ sao cho $(n+add)$ % 3 == 0  và $(n – sub)$ % 3 == 0

```
add = 3 – n % 3
sub = n % 3
```
  Nếu $sub$ = 0 thì $sub$ cập nhập bằng 3


*   **Bước 3**: Duyệt mảng :

Đầu tiên tìm phần tử lớn hơn n bằng cách duyệt từ đầu mảng đến cuối mảng


```
For (i = 0 đến i < n, i++)
	If (Num[i] + add <10):
		Num[i] = Num[i] + add
		While (Num[i] + 3 < 10)
			Num[i] = Num[i] + 3
		Return Num
```
Nếu không tìm thấy tiếp tục tìm số nhỏ hơn n bằng cách duyệt theo chiều ngược lại


```
For (i = 0 đến i < n, i++)
	If (Num[i] - sub >=0):
		Num[i] = Num[i] – sub
		Return Num
```




# **COMPLEXITY EVALUATE**



*   Time complexity: O(n) với n là độ dài của chuỗi



# **PROGRAMMING**

````python
def Tinh(num):
    count_num = 0
    sum_num = 0
    a = num
    Num = []
    while a != 0:
        count_num = count_num + 1
        b = a % 10
        Num.insert(0, b)
        sum_num = sum_num + b
        a = a // 10

    if sum_num % 3 == 0:
        for i, so in enumerate(Num):

            if so + 3 < 10:
                while Num[i] + 3 < 10:
                    Num[i] = Num[i] + 3
                return Num
        return Num
    else:
        so_thieu = 3-sum_num % 3 #can them bao nhieu so de chia het cho 3
        for i, so in enumerate(Num):
            if so + so_thieu < 10:
                while Num[i] + so_thieu < 10:
                    Num[i] = Num[i] + so_thieu
                return Num
        Num2 = Num[::-1]
        for i, so in enumerate(Num2):
            if so - sum_num % 3 > 0:
                Num2[i] = Num2[i] - sum_num % 3
                return Num2[::-1]
num = 12345
print(num)
Num = Tinh(num)
so = 0
for i in Num:
    so = so*10 + i
print(so)
````
