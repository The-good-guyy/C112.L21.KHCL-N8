1.   Abstraction: Cho mng A gm n s nguyn dng t [a1 -> an]. Tm s k ln nht c th sao cho c t nht k s trong dy c gi tr ln hn k.
2.   Pattern recognition: 

        . Bi ton sp xp cc phn t trong mng

        . c im nhn dng: tm s k ln nht trong mng, sao cho c t nht
        k s c gi tr ln hn k trong dy s.
3.   Algorithm designed:


      *   Nhp gi tr ca n (1 <= n <= 5x10^5)

      *   Nhp cc gi tr ci ca mng A gm n phn t (0 <= ci <= 10^6 , i = 1 % n)

      *   Sp xp cc phn t trong mng theo gi tr tng dn

      *   Gn H_ind = n

      *   Chy vng lp for t i = 0 -> n
                . Nu A[i] >= H_ind, th thot khi vng lp
                . Ngc li th H_ind = H_ind - 1

      *   Xut gi tr ca H_ind




4.   Time complexity:
      .  phc tp ca chng trnh l O(n), vi n l s phn t ca mng
5.   Programming:




```python
n = int(input())
l = list(map(int,input().split()))
l.sort()
H_ind = n
for i in range(n):
    if l[i] >= H_ind:
        break
    H_ind -= 1
print(H_ind)
```
