1.   ABSTRACTION

  *   Nhập vào 4 dòng, mỗi dòng chứa 1 số nguyên a,b,c,d (0 < a,b < 10^5 ),(0 < c < d <= 10^5)
  *   Tìm số lần biến đổi phân số a/b thành c/d. Nếu không biến đổi được xuất 0, ngược lại xuất số lần biến đổi



2.   PATTERN RECOGNITION

  *   Bài toán kiểm tra phân số a/b có thể biến đổi thành phân số c/d hay không. Tính n = a*d - b*c, nếu n < 0 thực hiện biến đổi đến khi a/b = c/d dừng

1.   ALGORITHM DESIGN

  *   Chúng ta sẽ tìm ước chung lớn nhất (UCLN) của hai số a và b

              def UCLN(a,b):
                if (b == 0):
                return a
              return UCLN(b,a%b)

  *   Tìm giá trị của n = a*d = b*c

              def TinhN(a,b,c,d,n):
                n = a*d - b*c
              return n

  *   Kiểm tra giá trị của n, nếu n > 0 (a/b > c/d) phân số a/b không thể biến đổi thành c/d. Xuất ra giá trị 0, kết thúc.

  *    Ngược lại nếu n < 0. Thực hiện việc lặp tăng giá trị của a và b lên 1 đơn vị. Sau mỗi lần lặp chia a và b cho UCLN(a,b), thực hiện tăng giá trị của biến đếm lên 1. Sau đó so sánh giá trị của a/b và c/d. Nếu a/b = c/d dừng và xuất ra giá trị đếm. Nếu không bằng tiếp tục vòng lặp.

1.   COMPLEXITY EVALUATE


*   Độ phức tạp của thuật toán: O(n)



5.   PROGRAMING


















```python
def UCLN(a,b):
    if (b == 0):
        return a
    return UCLN(b,a%b)

dem = 0
n = int()
def TinhN(a,b,c,d,n):
    n = a*d - b*c
    return n
def PhanSo(a,b,c,d,dem):
    if (TinhN(a,b,c,d,n) < 0):
        while (a/b < c/d):
            a += 1
            b += 1
            uc = UCLN(a,b)
            a //= uc
            b //= uc
            dem += 1
        if(a*d == b*c):
            print(dem)
        else:
            print(0)
    else:
        print(0)

if __name__=="__main__":
    a = int(input())
    b = int(input())
    c = int(input())
    d = int(input())
    PhanSo(a,b,c,d,dem)
```
